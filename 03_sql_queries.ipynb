```python
conn = sqlite3.connect('telco_churn.db')

query2 = """
SELECT 
    Contract,
    COUNT(*) as total_customers,
    SUM(Churn) as churned_customers,
    ROUND(100.0 * SUM(Churn) / COUNT(*), 2) as churn_rate_percent
FROM customers
GROUP BY Contract
ORDER BY churn_rate_percent DESC
"""

result2 = pd.read_sql_query(query2, conn)
print("📋 CHURN RATE BY CONTRACT TYPE:")
print("=" * 50)
print(result2.to_string(index=False))
print("=" * 50)

conn.close()
```

```python
conn = sqlite3.connect('telco_churn.db')

query3 = """
SELECT 
    PaymentMethod,
    COUNT(*) as total_customers,
    SUM(Churn) as churned_customers,
    ROUND(100.0 * SUM(Churn) / COUNT(*), 2) as churn_rate_percent
FROM customers
GROUP BY PaymentMethod
ORDER BY churn_rate_percent DESC
"""

result3 = pd.read_sql_query(query3, conn)
print("💳 CHURN RATE BY PAYMENT METHOD:")
print("=" * 55)
print(result3.to_string(index=False))
print("=" * 55)

conn.close()
```

```python
conn = sqlite3.connect('telco_churn.db')

query4 = """
SELECT 
    COUNT(*) as high_risk_customers,
    SUM(Churn) as churned_customers,
    ROUND(100.0 * SUM(Churn) / COUNT(*), 2) as churn_rate_percent
FROM customers
WHERE Contract = 'Month-to-month' 
  AND PaymentMethod = 'Electronic check'
"""

result4 = pd.read_sql_query(query4, conn)
print("⚠️ HIGH-RISK CUSTOMER SEGMENT:")
print("=" * 45)
print(result4.to_string(index=False))
print("=" * 45)
print("\n💡 Insight: Month-to-month + Electronic check = Highest churn risk!")

conn.close()
```

```python
conn = sqlite3.connect('telco_churn.db')

query5 = """
SELECT 
    CASE 
        WHEN tenure <= 12 THEN '0-12 months'
        WHEN tenure <= 24 THEN '13-24 months'
        WHEN tenure <= 48 THEN '25-48 months'
        WHEN tenure <= 72 THEN '49-72 months'
        ELSE '73+ months'
    END as tenure_group,
    COUNT(*) as total_customers,
    SUM(Churn) as churned_customers,
    ROUND(100.0 * SUM(Churn) / COUNT(*), 2) as churn_rate_percent
FROM customers
GROUP BY tenure_group
ORDER BY churn_rate_percent DESC
"""

result5 = pd.read_sql_query(query5, conn)
print("⏰ CHURN RATE BY TENURE GROUP:")
print("=" * 55)
print(result5.to_string(index=False))
print("=" * 55)

conn.close()
```

```python
# Save all queries to a .sql file for your GitHub portfolio
sql_queries = """
-- ============================================
-- TELCO CUSTOMER CHURN ANALYSIS
-- SQL QUERIES FOR PORTFOLIO
-- ============================================

-- 1. Overall churn rate
SELECT 
    COUNT(*) as total_customers,
    SUM(Churn) as churned_customers,
    ROUND(100.0 * SUM(Churn) / COUNT(*), 2) as churn_rate_percent
FROM customers;

-- 2. Churn rate by contract type
SELECT 
    Contract,
    COUNT(*) as total_customers,
    SUM(Churn) as churned_customers,
    ROUND(100.0 * SUM(Churn) / COUNT(*), 2) as churn_rate_percent
FROM customers
GROUP BY Contract
ORDER BY churn_rate_percent DESC;

-- 3. Churn rate by payment method
SELECT 
    PaymentMethod,
    COUNT(*) as total_customers,
    SUM(Churn) as churned_customers,
    ROUND(100.0 * SUM(Churn) / COUNT(*), 2) as churn_rate_percent
FROM customers
GROUP BY PaymentMethod
ORDER BY churn_rate_percent DESC;

-- 4. High-risk customer segment
SELECT 
    COUNT(*) as high_risk_customers,
    SUM(Churn) as churned_customers,
    ROUND(100.0 * SUM(Churn) / COUNT(*), 2) as churn_rate_percent
FROM customers
WHERE Contract = 'Month-to-month' 
  AND PaymentMethod = 'Electronic check';

-- 5. Churn rate by tenure group
SELECT 
    CASE 
        WHEN tenure <= 12 THEN '0-12 months'
        WHEN tenure <= 24 THEN '13-24 months'
        WHEN tenure <= 48 THEN '25-48 months'
        WHEN tenure <= 72 THEN '49-72 months'
        ELSE '73+ months'
    END as tenure_group,
    COUNT(*) as total_customers,
    SUM(Churn) as churned_customers,
    ROUND(100.0 * SUM(Churn) / COUNT(*), 2) as churn_rate_percent
FROM customers
GROUP BY tenure_group
ORDER BY churn_rate_percent DESC;
"""

with open('churn_analysis_queries.sql', 'w') as f:
    f.write(sql_queries)

print("✅ SQL queries saved to 'churn_analysis_queries.sql'")
print("📁 This file is ready for your GitHub portfolio!")
```
